## Distribuições: Normal, Binomial, etc.

Distribuições de probabilidade são fundamentais na estatística, descrevendo como as probabilidades estão distribuídas em diferentes resultados em um experimento.

**Distribuição Normal**

A distribuição normal, ou Gaussiana, é uma das mais comuns, caracterizada por sua forma de sino.

Características:

Simétrica em torno da média.
Definida por dois parâmetros: média (μ) e desvio padrão (σ).


```python
import numpy as np
import matplotlib.pyplot as plt

mu, sigma = 0, 0.1

s = np.random.normal(mu, sigma, 1000)

plt.hist(s, bins=30, density=True)

plt.show()
```

**Distribuição Binomial**

Descreve o número de sucessos em um número fixo de tentativas de um experimento binomial.

Características:

Os ensaios são independentes.
Cada ensaio tem apenas dois resultados possíveis: sucesso ou fracasso.

```python
n, p = 10, 0.5

binomial = np.random.binomial(n, p, 1000)

plt.hist(binomial, bins=30, density=False)

plt.xlabel('Número de sucessos')
plt.ylabel('Frequência')

plt.title('Distribuição Binomial')

plt.show()
```

Lançamento de Moeda

Sucesso: Um lançamento resulta em "cara". <br>
Fracasso: Um lançamento resulta em "coroa".

Neste caso:

( n ) é o número de lançamentos da moeda (ensaios).
( p ) é a probabilidade de obter "cara" em um único lançamento (probabilidade de sucesso).
Interpretação no Contexto do Exemplo
n = 10: Realizamos 10 lançamentos de moeda.
p = 0.5: A probabilidade de obter "cara" em um único lançamento é 0.5 (ou 50%).
O vetor binomial contém 1000 amostras, cada uma representando o número de "caras" (sucessos) em 10 lançamentos de moeda.

Sucesso: Obter "cara" em um lançamento.<br>
Fracasso: Obter "coroa" em um lançamento.<br>

Histograma <br>
O eixo x mostra o número de sucessos (quantas "caras" foram obtidas) em 10 lançamentos.
O eixo y mostra a contagem de quantas vezes cada número de sucessos ocorreu nas 1000 simulações.
Por exemplo, se uma barra tem altura 100 para o valor 5 no eixo x, isso significa que, em 1000 amostras, houve 100 simulações onde exatamente 5 "caras" foram obtidas em 10 lançamentos.

**Distribuição Poisson**

Descreve o número de eventos ocorrendo em um intervalo fixo de tempo ou espaço.

```python
lam = 5
poisson = np.random.poisson(lam, 1000)

plt.hist(poisson, bins=30, density=False)
plt.show()
```

**Distribuição Exponencial**

Descreve o tempo até o próximo evento em um processo de Poisson.

```python
scale = 1.0

exponencial = np.random.exponential(scale, 1000)


plt.hist(exponencial, bins=30, density=True)
plt.show()
```

Aplicações <br>
Normal: Usada em estatísticas inferenciais, finanças, ciência de dados, etc.<br>
Binomial: Aplicada para modelar eventos com dois possíveis resultados. <br>
Poisson e Exponencial: Utilizadas para modelar eventos que ocorrem aleatoriamente ao longo do tempo ou espaço.
